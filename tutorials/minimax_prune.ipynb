```python
from graphviz import Digraph
from IPython.display import display
```

```python
# Create the Digraph object
dot = Digraph()
dot.attr(rankdir='BT')

# Level 4 (Leaf nodes)
dot.node('H', '-1', style='filled', fillcolor='lightblue2')
dot.node('I', '3', style='filled', fillcolor='lightblue2')
dot.node('J', '5', style='filled', fillcolor='lightblue2')
dot.node('K', '1', style='filled', fillcolor='lightblue2')
dot.node('L', '-6', style='filled', fillcolor='lightblue2')
dot.node('M', '-4', style='filled', fillcolor='lightblue2')
dot.node('N', '0', style='filled', fillcolor='lightblue2')
dot.node('O', '9', style='filled', fillcolor='lightblue2')

# Level 3
dot.node('D', '3\nmax(-1,3)', style='filled', fillcolor='white')
dot.node('E', '5\nmax(5,1)', style='filled', fillcolor='white')
dot.node('F', '-4\nmax(-6,-4)', style='filled', fillcolor='white')
dot.node('G', '9\nmax(0,9)', style='filled', fillcolor='white')

# Level 2
dot.node('B', '3\nmin(3,5)', style='filled', fillcolor='lightblue2')
dot.node('C', '-4\nmin(-4,9)', style='filled', fillcolor='lightblue2')

# Level 1
dot.node('A', '3\nmax(3,-4)', style='filled', fillcolor='white')


# Add edges
# Level 3 to Level 4
dot.edge('H', 'D')
dot.edge('I', 'D')
dot.edge('J', 'E')
dot.edge('K', 'E')
dot.edge('L', 'F')
dot.edge('M', 'F')
dot.edge('N', 'G')
dot.edge('O', 'G')

# Level 2 to Level 3
dot.edge('D', 'B')
dot.edge('E', 'B')
dot.edge('F', 'C')
dot.edge('G', 'C')

# Level 1 to Level 2
dot.edge('B', 'A')
dot.edge('C', 'A')
```
